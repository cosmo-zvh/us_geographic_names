```python
import pandas as pd 

input_directory = "/kaggle/input"
output_directory = "/kaggle/working"
geo_names_us = "/usa-board-on-geographic-names/NationalFile_20210825.txt"
us_geo_names_input = input_directory + geo_names_us

```

Importing dataset to a Pandas Dataframe 

```python
column_names = [
    "feature_id",
    "feature_name",
    "feature_class",
    "state_alpha",
    "state_numeric",
    "county_name",
    "county_numeric",
    "primary_lat_dms",
    "prim_long_dms",
    "prim_lat_dec",
    "prim_long_dec",
    "source_lat_dms",
    "source_long_dms",
    "source_lat_dec",
    "source_long_dec",
    "elev_in_m",
    "elev_in_ft",
    "map_name",
    "date_created",
    "date_edited"
    ]

column_dtypes = {
    "feature_id": float,
    "feature_name": str,
    "feature_class": str,
    "state_alpha": str,
    "state_numeric": float,
    "county_name": str,
    "county_numeric": float,
    "primary_lat_dms": str,
    "prim_long_dms": str,
    "prim_lat_dec": float,
    "prim_long_dec": float,
    "source_lat_dms": str,
    "source_long_dms": str,
    "source_lat_dec": float,
    "source_long_dec": float,
    "elev_in_m": float,
    "elev_in_ft": float,
    "map_name": str,
    "date_created": str,
    "date_edited": str    
    }
    
us_geo_names_df = pd.read_csv(us_geo_names_input, 
                              sep="|", 
                              header=0, 
                              names=column_names,
                              index_col="feature_id",
                              parse_dates = [-1, -2],
                              dtype=column_dtypes,
                              )


display(us_geo_names_df.info())
display(us_geo_names_df.head())
```

Use the following if you want to generate a CSV

```python
us_geo_names_df.to_csv("us-geographic-names.zip", index=False)
```
